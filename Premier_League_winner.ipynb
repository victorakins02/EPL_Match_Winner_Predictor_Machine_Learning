```python
matches["date"] = pd.to_datetime(matches["date"])
# Changed to numerical vale we can now use
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
# Changing venue from string into a category then into numerical values.
# Trying to look into home/away advantage
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
# Assign each opponent a value
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex = True).astype("int")
# what hour did each team play
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
# Assign numerical values to the days
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
# 1 if team won, 0 if draw/loss
```

```python
from sklearn.ensemble import RandomForestClassifier
# Picks up non-linearity in data
# "opp_code" doesn't have a linear relationship
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
# n_estimators is number of decision trees we want to train. The more trees, the more accurate
# randomforest is a series of decision trees but each tree has different parameters
# min_samples_split is the number of samples before we split node. Higher value means likely to overfit, but lower accuary on training data.
# randomforest has lots of random parameters, so if we re-run the randomforest, we get the same results.
```

```python
train = matches[matches["date"] < '2022-01-01']
# our training data is 2021 season
```

```python
test = matches[matches["date"] > '2022-01-01']
# our test data is 2022 season
```

```python
predictors = ["opp_code", "venue_code", "hour", "day_code"]
# we will try predict the winner of each game based of these parameters
```

```python
rf.fit(train[predictors], train["target"])
# use our Randomforest to train our model.
```

```python
preds = rf.predict(test[predictors])
# Now we plan to compare again our test data
```

```python
from sklearn.metrics import accuracy_score
# importing accuracy_score to test the accuracy of our training data
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
acc
# we are 60% accurate
```

```python
combined = pd.DataFrame(dict(actual = test["target"], prediction=preds))
#Looking for when our accuracy is high or low.
#Data frame combines out actual and predicted values
```

```python
pd.crosstab(index=combined["actual"], columns = combined["prediction"])
# what happened when we predicted a 0 or 1
# we were right about the loses 146 times and wrong 82 times. This is accurate
# we were right about the wins 26 times and wrong 22 times. This is not accurate
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["target"], preds)
# new metric
# when we predicted a win, what percentage of the time did the team actually win
# so we only predicted correct wins 46% of the time
# how can we improve
```

```python
def rolling_averages(group, cols, new_cols):
  group = group.sort_values("date") # sort by date
  rolling_stats = group[cols].rolling(3, closed = 'left').mean() #rolling avg
  group[new_cols] = rolling_stats # assign back to dataframe with new name
  group = group.dropna(subset=new_cols) # drop missing values
  return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
# goals for, goals against, shots takens, shots on target, distance traveled, free kicks, penalty kick, penalty kick attempts
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
#collected rolling averages of each stat
```

```python
matches_rolling = matches.groupby("team").apply(lambda x : rolling_averages(x, cols, new_cols))
# first group everything by team
# apply the function to each team in dataset to get rolling averages of each team
```
