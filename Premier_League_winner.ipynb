```python
matches.dtypes
# For machiene learning, we need numerical data.
```

```python
matches["date"] = pd.to_datetime(matches["date"])
# Changed to numerical vale we can now use
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
# Changing venue from string into a category then into numerical values.
# Trying to look into home/away advantage
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
# Assign each opponent a value
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex = True).astype("int")
# what hour did each team play
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
# Assign numerical values to the days
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
# 1 if team won, 0 if draw/loss
```

```python
from sklearn.ensemble import RandomForestClassifier
# Picks up non-linearity in data
# "opp_code" doesn't have a linear relationship
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
# n_estimators is number of decision trees we want to train. The more trees, the more accurate
# randomforest is a series of decision trees but each tree has different parameters
# min_samples_split is the number of samples before we split node. Higher value means likely to overfit, but lower accuary on training data.
# randomforest has lots of random parameters, so if we re-run the randomforest, we get the same results.
```

```python
train = matches[matches["date"] < '2022-01-01']
# our training data is 2021 season
```

```python
test = matches[matches["date"] > '2022-01-01']
# our test data is 2022 season
```

```python
predictors = ["opp_code", "venue_code", "hour", "day_code"]
# we will try predict the winner of each game based of these parameters
```

```python
rf.fit(train[predictors], train["target"])
# use our Randomforest to train our model.
```

```python
preds = rf.predict(test[predictors])
# Now we plan to compare again our test data
```

```python
from sklearn.metrics import accuracy_score
# importing accuracy_score to test the accuracy of our training data
```

```python
acc = accuracy_score(test["target"], preds)
```
